```python
%matplotlib inline
#%matplotlib widget

import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']
```

```python
folder = '/home/hugens/shared/uni/project/tweedledum/examples/data'
batches = []
for filename in os.listdir(folder):
    filename = folder + '/' + filename
    if '03-07' in filename or '03-08' in filename:
        batches.append(np.loadtxt(filename, skiprows=1, dtype=int))

data = batches[0]
for batch in batches[1:]:
    data = np.concatenate([data, batch])
    
df = pd.DataFrame(data=data, columns=["N", 
                                      "K", 
                                      "m", 
                                      "mean_num_nodes", 
                                      "max_num_nodes", 
                                      "mean_num_edges", 
                                      "max_num_edges"])
df.m /= df.N
```

```python
fig, axs = plt.subplots(figsize=(10, 5), nrows=1, ncols=2, constrained_layout=True)

for i, N in enumerate(df.N.unique()):
    N_df = df.loc[df['N'] == N]
    axs[0].plot(N_df.m, N_df.mean_num_nodes, color=colors[i], label=f'N = {N}')
    axs[0].plot(N_df.m, N_df.max_num_nodes, ls='--', color=colors[i])
    axs[1].plot(N_df.m, N_df.mean_num_edges, color=colors[i], label=f'N = {N}')
    axs[1].plot(N_df.m, N_df.max_num_edges, ls='--', color=colors[i])
axs[0].set_xlabel('M/N')
axs[0].set_ylabel('num_qubits')
axs[1].set_xlabel('M/N')
axs[1].set_ylabel('depth')
axs[0].legend()
axs[1].legend()
plt.savefig('everything_is_linear.png', dpi=300)
plt.show()
```

```python
def mean_evol(arr):
    n_points = len(arr)
    n_samples = [i for i in 2**np.arange(4, int(np.log2(n_points)))]
    true_mean = np.mean(arr)
    plot_points = np.zeros(len(n_samples)+1)
    for i, n in enumerate(n_samples):
        plot_points[i] = np.mean(arr[:n])
    plot_points[-1] = true_mean
    plot_points /= true_mean
    n_samples.append(n_points)
    return n_samples, plot_points

def max_evol(arr):
    n_points = len(arr)
    n_samples = [i for i in 2**np.arange(4, int(np.log2(n_points)))]
    true_max = np.max(arr)
    plot_points = np.zeros(len(n_samples)+1)
    for i, n in enumerate(n_samples):
        plot_points[i] = np.max(arr[:n])
    plot_points[-1] = true_max
    plot_points /= true_max
    #plot_points -= np.min(plot_points)
    n_samples.append(n_points)
    return n_samples, plot_points
```

```python
folder = '/home/hugens/shared/uni/project/tweedledum/examples/data'
files_labels = []
for filename in os.listdir(folder):
    if 'N=' in filename:
        files_labels.append((filename, filename[15:-4].replace("_", " ")))
        
fig, axs = plt.subplots(figsize=(20, 5), nrows=1, ncols=2, constrained_layout=True)


for filename, label in files_labels:
    nodes, edges = np.loadtxt(folder + '/' + filename, delimiter=',', skiprows=1).transpose()
    axs[0].plot(*mean_evol(nodes), '.-', label=label)
    axs[1].plot(*max_evol(nodes), '.-', label=label)

for ax in axs:
    ax.set_xscale('log', base=2)
    ax.legend()
```
